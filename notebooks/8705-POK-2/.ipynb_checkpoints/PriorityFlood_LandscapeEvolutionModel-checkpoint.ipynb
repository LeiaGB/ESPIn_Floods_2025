# Landscape evolution model with Priority flood and Space_v2
<hr>

The priority flood flow director is designed to calculate flow properties over large scale grids. In the following notebook we illustrate how the priority flood flow accumulator can be used to simulate landscape evolution using the SPAVE_V2 Landlab component

```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm

from landlab import RasterModelGrid, imshow_grid, imshowhs_grid
from landlab.components import PriorityFloodFlowRouter, SpaceLargeScaleEroder
```

Create raster grid

```python
mg = RasterModelGrid((100, 100), xy_spacing=10.0)
z = mg.add_zeros("topographic__elevation", at="node")
mg.at_node["topographic__elevation"][mg.core_nodes] += np.random.rand(
    mg.number_of_core_nodes
)

s = mg.add_zeros("soil__depth", at="node", dtype=float)
mg.at_node["soil__depth"][mg.core_nodes] += 0.5
mg.at_node["topographic__elevation"] += mg.at_node["soil__depth"]

# fr = FlowAccumulator(mg, flow_director='D8')
# df = DepressionFinderAndRouter(mg)

fr = PriorityFloodFlowRouter(mg, flow_metric="D8", update_flow_depressions=True)


ha = SpaceLargeScaleEroder(mg, K_sed=0.0001, K_br=0.0001, phi=0.3, H_star=1)

br = mg.at_node["bedrock__elevation"]
z = mg.at_node["topographic__elevation"]

space_dt = 100
```

```python
z_ori = np.array(z)
for i in tqdm(range(100)):
    # Uplift
    br[mg.core_nodes] += 0.001 * space_dt
    z[mg.core_nodes] = br[mg.core_nodes] + s[mg.core_nodes]
    fr.run_one_step()
    # df.map_depressions()
    ha.run_one_step(dt=space_dt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from landlab import RasterModelGrid
from landlab.components import PriorityFloodFlowRouter, SpaceLargeScaleEroder

# 1. Create a model grid
mg = RasterModelGrid((100, 100), xy_spacing=10.0)

# 2. Add topographic elevation: slope + noise
z = mg.add_zeros("topographic__elevation", at="node")
x, y = mg.node_x, mg.node_y
z += (mg.node_y.max() - y) * 0.01  # gentle north-to-south slope
z += np.random.rand(mg.size("node")) * 0.1  # small noise

# 3. Add soil depth and update elevation
s = mg.add_zeros("soil__depth", at="node", dtype=float)
s += 0.5  # uniform soil depth
z += s  # topographic elevation includes soil

# 4. Set up flow routing and SPACE erosion
fr = PriorityFloodFlowRouter(mg, flow_metric="D8", update_flow_depressions=True)

ha = SpaceLargeScaleEroder(
    mg,
    K_sed=5e-5,    # sediment erodibility
    K_br=1e-6,     # bedrock erodibility
    phi=0.35,      # porosity
    H_star=1.0     # characteristic thickness
)

# 5. Run the model loop
n_steps = 1000
dt = 100  # years

for i in range(n_steps):
    fr.run_one_step()
    ha.run_one_step(dt)

# 6. Plot final topography
plt.figure(figsize=(8, 6))
imshow = mg.plot.imshow("topographic__elevation", cmap="terrain")
plt.colorbar(imshow)
plt.title("Final Topographic Elevation")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm  # Progress bar
from landlab import RasterModelGrid
from landlab.components import PriorityFloodFlowRouter, SpaceLargeScaleEroder

# Create grid
mg = RasterModelGrid((100, 100), xy_spacing=10.0)
z = mg.add_zeros("topographic__elevation", at="node")

x = mg.node_x
y = mg.node_y
grid_width = x.max()
center_x = grid_width / 2

# --- Sculpted elevation profile ---
base_slope = (mg.node_y.max() - y) * 0.0005
central_valley = -0.005 * np.exp(-((x - center_x) ** 2) / (2 * (200 ** 2)))
left_dip = -0.002 * np.exp(-((x - 100) ** 2) / (2 * (50 ** 2)))
right_dip = -0.002 * np.exp(-((x - (grid_width - 100)) ** 2) / (2 * (50 ** 2)))
z += base_slope + central_valley + left_dip + right_dip
z += np.random.rand(mg.size("node")) * 0.005  # very low noise

# Add soil
s = mg.add_zeros("soil__depth", at="node", dtype=float)
s += 0.3
z += s

# --- Set Boundary Conditions ---
mg.set_closed_boundaries_at_grid_edges(True, False, True, False)  # Close top and bottom
bottom_row = mg.nodes_at_bottom_edge
bottom_center = bottom_row[len(bottom_row) // 2]
mg.status_at_node[bottom_center] = mg.BC_NODE_IS_FIXED_VALUE  # Single outlet

# Flow routing and erosion
fr = PriorityFloodFlowRouter(mg, flow_metric="D8", update_flow_depressions=True)
ha = SpaceLargeScaleEroder(
    mg,
    K_sed=5e-6,
    K_br=1e-6,
    phi=0.35,
    H_star=1.0
)

# Run simulation with progress bar
n_steps = 1000
dt = 100  # years
for _ in tqdm(range(n_steps), desc="Eroding landscape"):
    fr.run_one_step()
    ha.run_one_step(dt)

# Plot result
plt.figure(figsize=(8, 6))
imshow = mg.plot.imshow("topographic__elevation", cmap="terrain")
plt.colorbar(imshow)
plt.title("Single Catchment with Progress Bar")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from landlab import RasterModelGrid
from landlab.components import PriorityFloodFlowRouter, SpaceLargeScaleEroder
from landlab.plot import imshow_grid

# Create a vertical-elongated grid
mg = RasterModelGrid((100, 50), xy_spacing=10.0)
z = mg.add_zeros("topographic__elevation", at="node")
x, y = mg.node_x, mg.node_y
grid_height = y.max()
center_x = x.max() / 2

# Create convex-up river profile (quadratic slope)
main_slope = 15.0 * (1 - (y / grid_height))**2  # steep at top, flat at bottom

# Add central valley to guide river
valley = -4.0 * np.exp(-((x - center_x) ** 2) / (2 * (120 ** 2)))

# Add some side gradient (uplift) to funnel flow from sides
side_pull = 0.5 * np.exp(-((y - grid_height / 2) ** 2) / (2 * (200 ** 2))) * (
    np.exp(-((x - 50) ** 2) / (2 * 300 ** 2)) + np.exp(-((x - (x.max() - 50)) ** 2) / (2 * 300 ** 2))
)

# Final elevation
z += main_slope + valley + side_pull
z += np.random.normal(scale=0.05, size=z.size)  # gentle noise

# Add soil
s = mg.add_zeros("soil__depth", at="node", dtype=float)
s += 0.3
z += s

# Set all boundaries closed
mg.set_closed_boundaries_at_grid_edges(True, True, True, True)

# Open only bottom-center node
bottom_nodes = mg.nodes_at_bottom_edge
bottom_center = bottom_nodes[len(bottom_nodes) // 2]
mg.status_at_node[bottom_center] = mg.BC_NODE_IS_FIXED_VALUE

# Components
fr = PriorityFloodFlowRouter(mg, flow_metric="D8", update_flow_depressions=True)
ha = SpaceLargeScaleEroder(
    mg,
    K_sed=3e-6,
    K_br=1e-6,
    phi=0.4,
    H_star=1.0
)

# Run evolution with progress bar
n_steps = 1200
dt = 100
for _ in tqdm(range(n_steps), desc="Evolving convex-up catchment"):
    fr.run_one_step()
    ha.run_one_step(dt)

# Plot
imshow_grid(mg, "topographic__elevation", cmap="terrain")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from landlab import RasterModelGrid
from landlab.components import PriorityFloodFlowRouter, SpaceLargeScaleEroder
from landlab.plot import imshow_grid

def generate_terrain(size, roughness):
    """Diamond-square terrain generation."""
    data = np.zeros((size, size))
    size_m1 = size - 1

    # Initialize corners
    data[0, 0] = np.random.uniform()
    data[0, size_m1] = np.random.uniform()
    data[size_m1, 0] = np.random.uniform()
    data[size_m1, size_m1] = np.random.uniform()

    def divide(side_length):
        half = side_length // 2
        if half < 1:
            return

        # Square step
        for x in range(half, size_m1, side_length):
            for y in range(half, size_m1, side_length):
                avg = (
                    data[x - half, y - half]
                    + data[x + half, y - half]
                    + data[x - half, y + half]
                    + data[x + half, y + half]
                ) / 4.0
                data[x, y] = avg + np.random.uniform(-roughness, roughness)

        # Diamond step
        for x in range(0, size_m1 + 1, half):
            for y in range((x + half) % side_length, size_m1 + 1, side_length):
                avg = (
                    data[(x - half) % size_m1, y]
                    + data[(x + half) % size_m1, y]
                    + data[x, (y - half) % size_m1]
                    + data[x, (y + half) % size_m1]
                ) / 4.0
                data[x, y] = avg + np.random.uniform(-roughness, roughness)

        divide(half)

    divide(size_m1)
    return data

# --- Main execution with Landlab ---
terrain_size = 257  # Must be power of 2 plus 1
roughness = 0.5

terrain = generate_terrain(terrain_size, roughness)
terrain = gaussian_filter(terrain, sigma=2)  # Optional smoothing

# --- Load into Landlab Grid ---
mg = RasterModelGrid((terrain_size, terrain_size), xy_spacing=10.0)
z = mg.add_field("topographic__elevation", terrain.ravel(), at="node", units="m")

# Optional soil layer
s = mg.add_zeros("soil__depth", at="node", dtype=float)
s += 0.3
z += s

# --- Set boundary conditions ---
mg.set_closed_boundaries_at_grid_edges(True, True, True, True)

# Open central bottom outlet
bottom_nodes = mg.nodes_at_bottom_edge
center_outlet = bottom_nodes[len(bottom_nodes) // 2]
mg.status_at_node[center_outlet] = mg.BC_NODE_IS_FIXED_VALUE

# --- Components ---
fr = PriorityFloodFlowRouter(mg, flow_metric="D8", update_flow_depressions=True)
ha = SpaceLargeScaleEroder(
    mg, K_sed=2e-6, K_br=1e-6, phi=0.4, H_star=1.0
)

# --- Run simulation ---
n_steps = 1000
dt = 100

for _ in range(n_steps):
    fr.run_one_step()
    ha.run_one_step(dt)

# --- Plot ---
imshow_grid(mg, "topographic__elevation", cmap="terrain")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from landlab import RasterModelGrid
from landlab.components import PriorityFloodFlowRouter, SpaceLargeScaleEroder

# Grid setup
mg = RasterModelGrid((100, 100), xy_spacing=10.0)
z = mg.add_zeros("topographic__elevation", at="node")

# Coordinates
x = mg.node_x
y = mg.node_y
xc = mg.node_x.max() / 2  # grid center x

# Sculpted elevation: gentle bowl in center, elevated sides
center_valley = -0.003 * np.exp(-((x - xc) ** 2) / (2 * (1000 ** 2)))  # wide central trough
y_slope = (mg.node_y.max() - y) * 0.0005  # gentle N-S slope
noise = np.random.rand(mg.size("node")) * 0.01  # minimal noise

# Final topography
z += center_valley + y_slope + noise

# Add thin, consistent soil depth
s = mg.add_zeros("soil__depth", at="node", dtype=float)
s += 0.3
z += s

# Flow routing
fr = PriorityFloodFlowRouter(mg, flow_metric="D8", update_flow_depressions=True)

# SPACE component with low erodibility
ha = SpaceLargeScaleEroder(
    mg,
    K_sed=7e-6,    # low to moderate sediment erodibility
    K_br=1e-6,
    phi=0.35,
    H_star=1.0
)

# Run model
n_steps = 1000
dt = 100  # years

for i in range(n_steps):
    fr.run_one_step()
    ha.run_one_step(dt)

```

```python
# These are plotting functions


def plotting(
    grid, topo=True, DA=True, hill_DA=False, flow_metric="D8", hill_flow_metric="Quinn"
):
    if topo:
        azdeg = 200
        altdeg = 20
        ve = 1
        plt.figure()
        plot_type = "DEM"
        ax = imshowhs_grid(
            grid,
            "topographic__elevation",
            grid_units=("deg", "deg"),
            var_name="Topo, m",
            cmap="terrain",
            plot_type=plot_type,
            vertical_exa=ve,
            azdeg=azdeg,
            altdeg=altdeg,
            default_fontsize=12,
            cbar_tick_size=10,
            cbar_width="100%",
            cbar_or="vertical",
            bbox_to_anchor=[1.03, 0.3, 0.075, 14],
            colorbar_label_y=-15,
            colorbar_label_x=0.5,
            ticks_km=False,
        )
    if DA:
        # %% Plot first instance of drainage_area
        grid.at_node["drainage_area"][grid.at_node["drainage_area"] == 0] = (
            grid.dx * grid.dx
        )
        plot_DA = np.log10(grid.at_node["drainage_area"] * 111e3 * 111e3)

        plt.figure()
        plot_type = "Drape1"
        drape1 = plot_DA
        thres_drape1 = None
        alpha = 0.5
        ax = imshowhs_grid(
            grid,
            "topographic__elevation",
            grid_units=("deg", "deg"),
            cmap="terrain",
            plot_type=plot_type,
            drape1=drape1,
            vertical_exa=ve,
            azdeg=azdeg,
            altdeg=altdeg,
            thres_drape1=thres_drape1,
            alpha=alpha,
            default_fontsize=12,
            cbar_tick_size=10,
            var_name="$log^{10}DA, m^2$",
            cbar_width="100%",
            cbar_or="vertical",
            bbox_to_anchor=[1.03, 0.3, 0.075, 14],
            colorbar_label_y=+5,
            colorbar_label_x=0.5,
            ticks_km=False,
        )

        props = dict(boxstyle="round", facecolor="white", alpha=0.6)
        textstr = flow_metric
        ax.text(
            0.05,
            0.95,
            textstr,
            transform=ax.transAxes,
            fontsize=10,
            verticalalignment="top",
            bbox=props,
        )

    if hill_DA:
        # Plot second instance of drainage_area (hill_drainage_area)
        grid.at_node["hill_drainage_area"][grid.at_node["hill_drainage_area"] == 0] = (
            grid.dx * grid.dx
        )
        # plotDA = np.log10(grid.at_node["hill_drainage_area"] * 111e3 * 111e3)
        # plt.figure()
        # imshow_grid(grid, plotDA,grid_units=("m", "m"), var_name="Elevation (m)", cmap='terrain')

        plt.figure()
        plot_type = "Drape1"
        # plot_type='Drape2'
        drape1 = np.log10(grid.at_node["hill_drainage_area"])
        thres_drape1 = None
        alpha = 0.5
        ax = imshowhs_grid(
            grid,
            "topographic__elevation",
            grid_units=("deg", "deg"),
            cmap="terrain",
            plot_type=plot_type,
            drape1=drape1,
            vertical_exa=ve,
            azdeg=azdeg,
            altdeg=altdeg,
            thres_drape1=thres_drape1,
            alpha=alpha,
            default_fontsize=10,
            cbar_tick_size=10,
            var_name="$log^{10} DA, m^2$",
            cbar_width="100%",
            cbar_or="vertical",
            bbox_to_anchor=[1.03, 0.3, 0.075, 14],
            colorbar_label_y=-15,
            colorbar_label_x=0.5,
            ticks_km=False,
        )

        props = dict(boxstyle="round", facecolor="white", alpha=0.6)
        textstr = hill_flow_metric
        ax.text(
            0.05,
            0.95,
            textstr,
            transform=ax.transAxes,
            fontsize=10,
            verticalalignment="top",
            bbox=props,
        )
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

from landlab import HexModelGrid, RasterModelGrid
from landlab.components import (
    DepressionFinderAndRouter,
    FlowAccumulator,
    FlowDirectorD8,
    FlowDirectorSteepest,
)
from landlab.plot.drainage_plot import drainage_plot


# create a plotting routine to make a 3d plot of our surface.
def surf_plot(mg, surface="topographic__elevation", title="Surface plot of topography"):
    plt.figure()
    ax = plt.axes(projection="3d")

    # Plot the surface.
    Z = mg.at_node[surface].reshape(mg.shape)
    color = cm.gray((Z - Z.min()) / (Z.max() - Z.min()))
    ax.plot_surface(
        mg.x_of_node.reshape(mg.shape),
        mg.y_of_node.reshape(mg.shape),
        Z,
        rstride=1,
        cstride=1,
        facecolors=color,
        linewidth=0.0,
        antialiased=False,
    )
    ax.view_init(elev=35, azim=-120)
    ax.set_xlabel("X axis")
    ax.set_ylabel("Y axis")
    ax.set_zlabel("Elevation")
    plt.title(title)
    plt.show()
```

```python
mg = RasterModelGrid((10, 10))
_ = mg.add_field(
    "topographic__elevation", 3.0 * mg.x_of_node**2 + mg.y_of_node**2, at="node"
)
surf_plot(mg, title="Grid 1")
```
